```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

X = df[[
    'Year',
    'Duration_Days',
    'Military_Deaths_A',
    'Military_Deaths_B',
    'GDP_B_Billions',
    'Air_Strikes'
]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# criar dataframe
dados = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
dados['Naval_Battles'] = df['Naval_Battles'].values

# conferir
print(dados.describe())
print(pca.explained_variance_ratio_)
```

```python
#plot 2d com a naval battles
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(dados['PC1'], dados['PC2'], c=dados['Naval_Battles'], cmap='viridis', alpha=0.7)
plt.colorbar(label='Naval Battles')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Naval Battles')
plt.show()

```
