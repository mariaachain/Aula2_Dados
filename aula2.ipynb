Desafio Final Sci Tec
1. Crie uma lista chamada valores_brutos com os números: [10, 50, 150, 20, 300]
2. Crie uma função com Type Hints que percorra essa lista
3. Se o valor for maior que 100, coloque-o em uma nova lista chamada alerta
4. NO final, use um if para printar: "SIStema seguro" (se não houver alertas) ou " !! Revisar valores: {lista}" (se houver)

```python
# 1. Definição dos dados iniciais (Casting e Type Hints)
valores_brutos: list = [10, 50, 150, 20, 300]

# 2. Criando função com Type Hints
def analis_integ(dados: list) -> list:
    '''
    recebe lista de valores e imprime apenas os acima de 100'''

    #Usando list comprehension (Intermediario)
    alertas =  [valor for valor in dados if valor > 100]
    return alertas

# 3. Executando lógica 
lista_de_alertas = analis_integ(valores_brutos)

# 4. Estrutura de decisão 
if not lista_de_alertas:
    print(" :) Sistema seguro: Nenhum valor crítico encontrado! ")
else:
    print(f" :( Atenção SciTec! Sistema em alerta, revisar seguintes valores: {lista_de_alertas}")
```
